```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets  import make_blobs
```

```python
X,y=make_blobs(n_samples=500,n_features=2,centers=5,random_state=3)
X
```

```python
plt.figure(0)
plt.grid(True)
plt.scatter(X[:,0],X[:,1],c=y)
plt.show()
```

```python
k=5
color=['green','yellow','blue','cyan','red']
clusters={}
for idx in range(k):
    center=10*(2*np.random.random((X.shape[1],))-1)
    point=[]
    cluster={
        'center':center,
        'points':point,
        'color':color[idx]


    }
    clusters[idx]=cluster
```

```python
clusters
```

```python
plt.scatter(X[:,0],X[:,1],c=y)
for i in clusters:
    center=clusters[i]['center']
    plt.scatter(center[0],center[1],marker='^',c='red')
plt.show()
```

```python
def distance(v1,v2):
    return(np.sqrt(np.sum(v1-v2)**2))
```

```python
def assighn():
    for i in range(X.shape[0]):
        dist=[]
        curr_id=X[i]
        # print(curr_id)
        for i in range(k):
            dis=distance(curr_id,clusters[i]['center'])
            dist.append(dis)
        cur_cluster=np.argmin(dist)
        clusters[cur_cluster]['points'].append(curr_id)
```

```python

def points_assign():
    for i in clusters:
        pts = clusters[i]['points']
        try:
            pts=np.array(pts)
            plt.scatter(pts[:, 0], pts[:, 1], c=clusters[i]['color'])
        except Exception as e:
            print(f"Error plotting points for cluster {i}: {e}")
        center = clusters[i]['center']
        plt.scatter(center[0], center[1], c='black', marker='^')

    plt.show()
```

```python
def updatecluster():
    for i in clusters:
        pts=np.array(clusters[i]['points'])
        if pts.shape[0]>0:
            new_center=pts.mean(axis=0)
            clusters[i]['center']=new_center
            clusters[i]['points']=[]
```

```python
assighn()
points_assign()
updatecluster()
```
